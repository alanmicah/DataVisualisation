```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
flight_filepath = ('/Users/micah/build/DataVisualisation/input/flight_delays.csv')
flight_data = pd.read_csv(flight_filepath, index_col='Month')
```

```python
flight_data
```

```python
# Set the width and height of graph
plt.figure(figsize=(10,6))

# Add title
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

# Bar chart showing average arrival delay for Spirit Airlines flights by month
# sns.barplot tells notebook we want to create a bar chart
# x= determines what to use on horizontal axis (in this case index is used which is the months)
# y= determines what to use on vertical axis (in this case column 'NK' which is Spirit Airlines)
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```
