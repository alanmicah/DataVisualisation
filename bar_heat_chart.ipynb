```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
flight_filepath = ('/Users/micah/build/DataVisualisation/input/flight_delays.csv')
flight_data = pd.read_csv(flight_filepath, index_col='Month')
```

```python
flight_data
```

```python
# Set the width and height of graph
plt.figure(figsize=(10,6))

# Add title
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')

# Bar chart showing average arrival delay for Spirit Airlines flights by month
# sns.barplot tells notebook we want to create a bar chart
# x= determines what to use on horizontal axis (in this case index is used which is the months)
# y= determines what to use on vertical axis (in this case column 'NK' which is Spirit Airlines)
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

```python
plt.figure(figsize=(10,6))
plt.title('Average Arrival Delay for Spirit Airlines Flights, by Month')
# Heatmap showing average arrival delay for each airline by month
# annot= write data in each cell
sns.heatmap(data=flight_data, annot=True)
plt.xlabel('Airline')
```

```python
# Exercise
ign_filepath = ('/Users/micah/build/DataVisualisation/input/ign_scores.csv')
ign_data = pd.read_csv(ign_filepath, index_col='Platform')
ign_data
```

```python
# Highest average score for PC for any genre
def hi_lo_average(platform_loc, data):
  a = 0
  b = 10
  for n, i in enumerate(data.loc[platform_loc]):
    if i > a:
      a = i
      index_hi = data.loc[platform_loc].index[n]
    if i < b:
      b = i
      index_low = data.loc[platform_loc].index[n]
  high_score = a
  worst_genre = b
  return high_score, worst_genre, index_hi, index_low

print(hi_lo_average('PC', ign_data))

# for i in ign_data:
#   print(i)
# for genre_name, genre in ign_data.items():
#   print(genre_name, genre)
#   # ign_data.loc[genre]
#   None

# Lowest total average score across all the Platforms
row_average = 100
low_index = ""
for index, row in ign_data.iterrows():
  c = row.sum() / len(ign_data)
  if c < row_average:
    row_average = c
    low_index = index
print(low_index)
```
