```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
cancer_filepath = '/Users/micah/build/DataVisualisation/input/cancer.csv'
cancer_data = pd.read_csv(cancer_filepath, index_col='Id')
cancer_data.head()
```

```python
# Find largest value in 'Perimeter (mean)' for the first 5 rows
x = 0
for n, i in enumerate(cancer_data['Perimeter (mean)'][:5]):
  if i > x:
    x=i
max_perim = x
print(max_perim)

# Find by row id
print(cancer_data.loc[8510824]['Perimeter (mean)'])
```

```python
# Two histograms showing distribution in values for 'Area (mean)'
# , separately for both benign and malignant tumors.
sns.histplot(data=cancer_data, x='Area (mean)', hue='Diagnosis')
plt.title('benign and maligant tumors')

# Malignant tumors have higher values for 'Area (mean)', on average relative to Bengn tumors.
# Malignant tumors also have a larger range of potential values
```

```python
# KDE plot
sns.kdeplot(data=cancer_data, x='Radius (worst)', hue='Diagnosis', fill=True)
# Given a tumor with a value for 'Radius (worst)' of 25
#, the algorithm will more likely classify the tumor as Malignant,
#, as the curve for Malignant at 25 is much higher
```
